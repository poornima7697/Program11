```python
dnames=['Satara','pune','mumbai',"nashik"]
pOfE=[60,150,200,70]
pOfW=[34,20,250,170]
pOfS=[60,160,100,200]
pOfN=[60,450,200,100]
plt.plot(dnames,pOfE,"g*--",markeredgecolor='red',label="East")
plt.plot(dnames,pOfW,color='red',label='west')
plt.plot(dnames,pOfN,color='yellow',label='south')
plt.plot(dnames,pOfW,color='blue',label='north')
plt.xlabel("District")
plt.ylabel("population")
plt.legend(bbox_to_anchor=(0.70,1.3),ncol=2)
plt.show()
```

```python
dnames=['Satara','pune','mumbai',"nashik"]
pOfE=[60,150,200,70]
pOfW=[34,20,250,170]
pOfS=[60,160,100,200]
pOfN=[60,450,200,100]
plt.xlabel("District")
plt.ylabel("population")
plt.title("population of district")
plt.legend(bbox_to_anchor=(0.70,1.3),ncol=2)
bar1=[2,4,6,8]
bar2=[]
plt.bar(bar1,pOfE,width=0.4,label="EastRegion")
for i in bar1:
    bar2.append(i+0.4)
plt.bar(bar2,pOfW,width=0.4,label="WestRegion")
plt.show()
```

```python
dnames=['Satara','pune','mumbai',"nashik"]
Day1=[60,150,200,70]
Day2=[34,20,250,170]
Day3=[60,160,100,200]
plt.xlabel("District")
plt.ylabel("temprature")
plt.title("temprature of district")
plt.legend(bbox_to_anchor=(0.70,1.3),ncol=3)
bar1=[2,4,6,8]
bar2=[]
plt.bar(bar1,Day1,width=0.4,label="Day1Region")
for i in bar1:
    bar2.append(i+0.4)
plt.bar(bar2,Day2,width=0.4,label="Day2Region")
bar3=[]
for i in bar2:
    bar3.append(i+0.4)
plt.bar(bar3,Day3,width=0.4,label="Day3Region")
plt.xticks(bar2,dnames)
plt.show()
```

```python
#stacked Bar plot
dnames=['Satara','pune','mumbai',"nashik"]
pOfE=[60,150,200,70]
pOfW=[34,20,250,170]
pOfS=[60,160,100,200]
plt.xlabel("District")
plt.ylabel("temprature")
plt.title("temprature of district")
plt.bar(dnames,pOfE,color='yellow',width=0.4,label="EastRegion")
plt.bar(dnames,pOfW,bottom=pOfE,color='green',width=0.4,label="WestRegion")
bot3=[i+j for i,j in zip(pOfE,pOfW)]
plt.bar(dnames,pOfS,bottom=bot3,color='magenta',width=0.4,label="south3Region")
plt.legend(bbox_to_anchor=(0.98,1.2),ncol=3)
#plt.xticks(bar2,dnames)
plt.show()
```
